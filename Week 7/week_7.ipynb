```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('view.jpg')
hsvImage = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

plt.figure(figsize=(10, 5))


plt.imshow(hsvImage)
plt.title("RGB to HSI Image")
```

```python
rgbImage = cv2.cvtColor(hsvImage, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(10, 5))


plt.imshow(rgbImage)
plt.title("HSI to RGB Image")
```
