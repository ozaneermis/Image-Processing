```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
image = cv2.imread('akciger.png')
kernel1 = np.ones((5, 5), np.float32)/30
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)

res = np.hstack((image,img))
plt.imshow(res)
plt.show()
```

```python
img = cv2.imread("test5.png")
img1 = cv2.blur(img,(5,5))
img2 = cv2.boxFilter(img,-1,(2,2), normalize=True)
img3 = cv2.medianBlur(image,5)
compare = np.concatenate((img1, img2,img3), axis=1)
plt.imshow(compare)
plt.show()
```
